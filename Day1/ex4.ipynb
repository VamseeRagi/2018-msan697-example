```python
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
```

```python
lines = sc.textFile("../Data/filtered_registered_business_sf.csv")
sf_business = lines.map(lambda x: (x.split(",")[0],x))
```

```python
sf_business.glom().first()
```

```python
lines = sc.textFile("../Data/supervisor_sf.csv")
supervisor = lines.map(lambda x: (x.split(",")[0],x))
```

```python
import time
```

```python
#without partitioning
start = time.time()
sf_business.join(supervisor).collect()
print time.time() - start
```

```python
partition_size = 6
sf_business = sf_business.partitionBy(partition_size).persist()
supervisor = supervisor.partitionBy(partition_size).persist()
```

```python
sf_business.take(1)
supervisor.take(1)
```

```python
#with partitioning
start = time.time()
sf_business.join(supervisor).collect()
print time.time() - start
```

```python
sf_business.join(supervisor).glom().collect()
```
